```python
%matplotlib inline

def show_plot(fn):
    from matplotlib import pylab as plt
    xs = [x for x in range(0,15)]
    ys = [fn(x) for x in xs]
    plt.plot(xs,ys)
    plt.show()
    
```

```python
import scipy.linalg
from numpy import sin, exp

f_x = lambda x: sin(x / 5.0) * exp(x / 10.0) + 5 * exp(-x / 2.0)

def buld_aproximate_fn(res):
    return lambda x : sum(c* (x ** i) for i, c in enumerate(res))
                          
show_plot(f_x)
```

```python
polynom_rank = 1
A_1 = [
    [1 ** n for n in range(0, polynom_rank + 1)],
    [15 ** n for n in range(0, polynom_rank + 1)]
]
b_1 = [f_x(1), f_x(15)]
show_plot(buld_aproximate_fn(scipy.linalg.solve(A_1, b_1)))
```

```python
polynom_rank = 2
A_2 = [
    [1 ** n for n in range(0, polynom_rank + 1)],
    [8 ** n for n in range(0, polynom_rank + 1)],
    [15 ** n for n in range(0, polynom_rank + 1)]
]

b_2 = [f_x(1), f_x(8), f_x(15)]

show_plot(buld_aproximate_fn(scipy.linalg.solve(A_2, b_2)))
```

```python
polynom_rank = 3
A_3 = [
    [1 ** n for n in range(0, polynom_rank + 1)],
    [4 ** n for n in range(0, polynom_rank + 1)],
    [10 ** n for n in range(0, polynom_rank + 1)],
    [15 ** n for n in range(0, polynom_rank + 1)]
]

b_3 = [f_x(1), f_x(4), f_x(10), f_x(15)]

show_plot(buld_aproximate_fn(scipy.linalg.solve(A_3, b_3)))
```

```python
for polynom_rank in range(1, 15):
  A_n = [[r ** n for n in range(0, polynom_rank + 1)] for r in range(1,polynom_rank+2)]
  b_n = [f_x(r) for r in range(1,polynom_rank+2)]
  show_plot(buld_aproximate_fn(scipy.linalg.solve(A_n, b_n)))
```
